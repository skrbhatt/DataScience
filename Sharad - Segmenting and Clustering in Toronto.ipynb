### Read the page and add the entries into a dataframe

```python
html = wp.page("List of postal codes of Canada:_M").html().encode("UTF-8")
```

```python
df = pd.read_html(html, header = 0)[0]
```

### This dataframe now has the raw data. It contains 'Not Assigned' in both features 'Borough' and 'Neighbourhood'

```python
df
```

### Delete/ignore the rows where 'Borough' is not assigned

```python
df = df[df.Borough != 'Not assigned']
```

```python
df
```

```python
df = df.groupby(['Postcode', 'Borough'])['Neighbourhood'].apply(list).apply(lambda x:', '.join(x)).to_frame().reset_index()
```

```python
df
```

### copy the Borough name to Neighbourhood, when a 'not assigned' is found in the 'Neighbourhood'

```python
for index, row in df.iterrows():
    if row['Neighbourhood'] == 'Not assigned':
        row['Neighbourhood'] = row['Borough']
```

```python
df_latlong = pd.read_csv("http://cocl.us/Geospatial_data")
```

```python
df_latlong.set_index('Postal Code')
```

### changing the column name 'Postcode' to match in both data frames

```python
df_latlong.columns = ['Postcode', 'Latitude', 'Longitude']
```

```python
df_latlong.head()
```

### Merge the two data frames into a new dataframe - df_final, using the primary key Postcode

```python
df_final = pd.merge(df, df_latlong[['Postcode','Latitude','Longitude']], on = 'Postcode')
```
